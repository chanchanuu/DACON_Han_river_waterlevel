```python
import pandas as pd
```

```python
### 데이터 불러오기
data = pd.read_csv('rawdata.csv')


### '공용량', '현재수위' 교체
팔당댐 = pd.read_csv('팔당댐_수정.csv', encoding='cp949')
data[팔당댐.columns[1:3]] = 팔당댐.iloc[:, 1:3]

### '한강대교_유량', '청담대교_유량' 교체
한강대교_유량 = pd.read_csv('한강대교_유량.csv')
data['한강대교_유량'] = 한강대교_유량['한강대교_유량']

청담대교_유량 = pd.read_csv('청담대교_유량_0822.csv', encoding='cp949')
data['청담대교_유량'] = 청담대교_유량.iloc[:, 2]

### 수위 column 4개, '행주대교_유량' 교체
수위_행주유량 = pd.read_csv('수위4개_행주대교유량.csv', encoding='cp949')
data[수위_행주유량.columns[1:]] = 수위_행주유량.iloc[:, 1:]


### '방류량_시간차' 추가
방류량_시간차 = pd.read_csv('방류량_시간차_수정.csv', encoding='cp949')
data['방류량_시간차'] = 방류량_시간차

### '강화대교_조위_시간차' 추가
강화대교_시간차 = pd.read_csv('강화대교조위_시간차.csv', encoding='cp949')
data['강화대교_조위_시간차'] = 강화대교_시간차

### '서울시 강수량' 추가
서울시강수량 = pd.read_csv('서울시강수량.csv')
data['서울시강수량'] = 서울시강수량


### 불필요한 column 버리기
data.drop(['유입량', '저수량', '방류량', '잠수교_유량', '대곡교_강수량', '진관교_강수량', '송정동_강수량', '강화대교_조위'],
          axis=1, inplace=True)

### Index 처리
data.set_index('날짜', drop=True, inplace=True)
data
```

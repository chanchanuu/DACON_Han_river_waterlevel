#시간차 적용을 위한 4월 데이터(조위df) 불러오기

```python
from glob import glob

w_list = sorted(glob("/content/drive/MyDrive/신입기수 프로젝트/Dataset/공용량_시간차_크롤링/*.csv"))
w_list
```

```python
idx = [0]
for i in range(47):
  idx.append(i)
```

```python
idx2 = []
for i in range(12):
  idx2.append(i*48)
```

```python
tmp3 = []

for i in range(11):
  a = pd.read_csv(w_list[i], encoding='cp949', usecols = [0,1])
  a.columns = [['날짜','공용량']]
  tmp3.append(a.iloc[idx].set_index(pd.Index(list(range(i+idx2[i],i+idx2[i]+48)))))

팔당df = pd.concat([tmp3[0],tmp3[1],tmp3[2],tmp3[3],tmp3[4],tmp3[5],tmp3[6],tmp3[7],tmp3[8],tmp3[9],tmp3[10]])
팔당df.columns = ['날짜', '공용량']
팔당df
```

```python
팔당df[팔당df['공용량']=='-']
```

#다리별 시간차 계산

```python
idx3 = []
for i in range(12):
  idx3.append(i*26496)
```

```python
ttmp = []
for i in range(0, 48): 
  
  push_num = i
  
  tmp = []
  for j in range(11):
    if j != 10:
      tmp.append(pd.concat([팔당df[idx2[j+1]-push_num:idx2[j+1]],팔당처리df[idx3[j]:idx3[j+1]-push_num][['날짜','공용량']]], ignore_index=True))
    else:
      tmp.append(pd.concat([팔당df[idx2[j+1]-push_num:idx2[j+1]],팔당처리df[idx3[j]:-push_num][['날짜','공용량']]], ignore_index=True))

  push_result = pd.concat(
        [tmp[0],tmp[1],tmp[2],tmp[3],tmp[4],tmp[5],tmp[6],tmp[7],tmp[8],tmp[9],tmp[10]], ignore_index=True
    )

  mm = pd.concat([push_result[['공용량']],팔당처리df[['청담대교_수위','잠수교_수위', '한강대교_수위', '행주대교_수위']]], axis=1)
  mm['공용량'] = mm['공용량'].astype(float, errors = 'raise')
  nn = mm[:269424]
  push_cor2 = nn.corr()
  
  
  print(i, push_cor2.iloc[:,0][2]) #마지막 원소 변경 1,2,3,4
  ttmp.append(push_cor2.iloc[:,0][2])

  #1번: 청담대교 0 0.16661086117544954
  #2번: 잠수교   0 0.19207065989114308
  #3번: 한강대교 0 0.19693234447794916
  #4번: 행주대교 0 0.16817380083101077
  #시간적 순서가 직관과 맞음

  #하나만 꼽으라면 19
  #print(i, push_cor2.sum()[0])
  #ttmp.append(push_cor2.sum()[0])
```
